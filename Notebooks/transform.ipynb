### Melt the dataframe to transform it


```python
# Melt the dataframe to transform it
df_melted = df.melt(id_vars=["Crop"], var_name="Date", value_name="Price")
```

```python
# Extract Year and Month from the column headers
df_melted["Month_Abbr"] = df_melted["Date"].str[:3]   # Extracts 'Jan', 'Feb', etc.
df_melted["Year"] = "20" + df_melted["Date"].str[-2:] # Converts '13' to '2013'
```

```python
import datetime
# Convert month abbreviation to full month name
df_melted["Month"] = df_melted["Month_Abbr"].apply(lambda x: datetime.datetime.strptime(x, "%b").strftime("%B"))

# Drop the temporary Month_Abbr column
df_melted.drop(columns=["Month_Abbr", "Date"], inplace=True)
```

```python
# Rearrange columns
df_melted = df_melted[["Crop", "Year", "Month", "Price"]]
df_melted
```

```python
# Save the output to a new CSV file
df_melted.to_csv("transformed_data.csv", index=False)
```
